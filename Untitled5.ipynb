```python
import pandas as pd

# Assuming 'summary_df' is the DataFrame with factual information
# and 'likert_columns' is a list of columns containing Likert responses


# Create a dictionary to store the grouped results
grouped_results = {}

# Loop through the Likert scale columns
for likert_column in likert_columns:
    grouped_results[likert_column] = {}

    # Loop through summary columns to group by (e.g., 'Gender', 'Department')
    for group_column in df.columns:
        if group_column not in likert_columns:  # Avoid grouping by the Likert columns
            # Get unique values in the current group column
            unique_groups = df[group_column].unique()

            # Loop through the unique groups in the current group column (e.g., Male, Female)
            for group_value in unique_groups:
                # Filter the DataFrame for the current group value
                filtered_df = df[df[group_column] == group_value]

                # Now, count the Likert scale responses for this specific group
                likert_counts = {}
                for option in ["Strongly Agree", "Agree", "Neutral", "Disagree", "Strongly Disagree"]:
                    count = filtered_df[likert_column].str.contains(option).sum() if option in filtered_df[likert_column].values else 0
                    likert_counts[option] = count

                # Store the Likert scale counts for the specific group
                grouped_results[likert_column][(group_column, group_value)] = likert_counts
```

```python
import pandas as pd
import os

# Assuming you have the 'grouped_results', 'likert_columns', and 'summary_df' from previous steps

# Get all unique group column names (excluding Likert columns)
group_columns = [col for col in df.columns if col not in likert_columns]

# Create a directory to store the CSV files (if it doesn't exist)
os.makedirs('grouped_likert_results', exist_ok=True)

# Loop through each group column
for group_column in group_columns:
    # Get unique values in the current group column
    unique_groups = df[group_column].unique()

    # Loop through each unique group value
    for group_value in unique_groups:
        # Create an empty list to store the data for the current DataFrame
        data = []

        # Loop through each Likert scale column
        for likert_column in likert_columns:
            # Get results for male and female for the current Likert column and group value
            male_results = grouped_results[likert_column].get(('Gender', 'Male'), {})  # Handle missing data
            female_results = grouped_results[likert_column].get(('Gender', 'Female'), {}) # Handle missing data

            # Create data rows for male and female (appending group_value for clarity)
            male_row = [likert_column, group_value, 'Male'] + list(male_results.values())
            female_row = [likert_column, group_value, 'Female'] + list(female_results.values())

            # Append rows to data
            data.extend([male_row, female_row])

        # Create column names
        columns = ['Likert Column', group_column, 'Gender'] + list(male_results.keys())

        # Create the DataFrame
        group_df = pd.DataFrame(data, columns=columns)

        # Export to CSV
        file_name = f'{group_column}_{group_value}.csv'
        file_path = os.path.join('grouped_likert_results', file_name)
        group_df.to_csv(file_path, index=False)
```

```python
import pandas as pd
import os
import itertools

# Assuming you have the 'grouped_results', 'likert_columns', and 'summary_df' from previous steps

# Get all unique group column names (excluding Likert columns)
group_columns = [col for col in df.columns if col not in likert_columns]

# Create a directory to store the CSV files (if it doesn't exist)
os.makedirs('grouped_likert_results', exist_ok=True)

# Generate all combinations of group column values
all_group_combinations = list(itertools.product(*[df[col].unique() for col in group_columns]))

# Loop through each combination of group values
for group_values in all_group_combinations:
    # Create a filter for the current combination of group values
    filter_condition = pd.Series(True, index=df.index)
    for i, group_column in enumerate(group_columns):
        filter_condition &= (df[group_column] == group_values[i])

    # Filter the DataFrame for the current group combination
    filtered_df = df[filter_condition]

    # Create an empty list to store the data for the current DataFrame
    data = []

    # Loop through each Likert scale column
    for likert_column in likert_columns:
        # Now, count the Likert scale responses for this specific group combination
        likert_counts = {}
        for option in ["Strongly Agree", "Agree", "Neutral", "Disagree", "Strongly Disagree"]:
            count = filtered_df[likert_column].str.contains(option).sum() if option in filtered_df[likert_column].values else 0
            likert_counts[option] = count

        # Append the results as a row to the data list
        data.append([likert_column] + list(group_values) + list(likert_counts.values()))

    # Create column names
    columns = ['Likert Column'] + group_columns + list(likert_counts.keys())

    # Create the DataFrame
    group_df = pd.DataFrame(data, columns=columns)

    # Generate a file name based on the group values
    file_name = '_'.join([f'{col}_{val}' for col, val in zip(group_columns, group_values)]) + '.csv'
    file_path = os.path.join('grouped_likert_results', file_name)

    # Export to CSV
    group_df.to_csv(file_path, index=False)
```
